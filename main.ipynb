```python
import numpy as np
```

### creating simple array

```python
# one dimention array
a=np.array([1,2,3])
# multi dimention array
b=np.array([[2,4,2],[5,2,4]])
print('this is one dimention array : \n',a) 
print('this is multi dimention array : \n',b) 
```

### basic info

```python
# finding the shape of matrix 
print(a.shape)
print(b.shape)
```

### reshape the matrix

```python
reshaped_a=a.reshape(3,1)
print(reshaped_a.shape)


reshaped_b=b.reshape(3,2)
print(reshaped_b.shape)
```

### if we dont know the shape and want to reshape just add -1 

```python
unknown_reshape=b.reshape(-1,3)
unknown_reshape.shape
```

### concatenate

```python
a = np.array([[3, 5], [6, 2]])
b = np.array([[7, 11]])
add_to_row=np.concatenate((a,b),axis=0)
add_to_col=np.concatenate((a,b.T),axis=1)
print(add_to_row)
print('\n')
print(add_to_col)
```

### Mathmathic operation 

 add a(n,n) to b(n,b)=c(n,n)

```python
a = np.array([1,2,3])
b = np.array([4,5,6])
c = a + b
print(c)
```

add a(1,n) or a(n,1) to b(n,m)=c(n,m)  


```python
a=np.array([1])
b=np.array([1,2,3])
c=a+b
print(c)

# for more details check numpy doc and search about broadcusting 
```

#### create array from scratch
